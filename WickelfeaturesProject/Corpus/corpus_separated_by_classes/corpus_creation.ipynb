```python
import pandas as pd
from glob import glob
import re
```

```python
df = pd.concat([pd.read_csv(f, header=None) for f in glob('train_*.csv')], ignore_index = True)
df
```

```python
df = pd.DataFrame()
for f in glob('train_*.csv'):
     #if f != 'train_polemicos.csv':
    temp = pd.read_csv(f, header=None)
    classe = re.search('_\w*', f)
    temp['class'] = 'class_' + classe.group()
    df = pd.concat([df, temp], ignore_index = True)
    
```

```python
df.columns = ['inf', 'v_inf', 'conj', 'v_conj', 'class']
```

```python
df = df.drop(['inf','conj'], axis=1)
```

```python
df["v_inf"] = df.v_inf.str.strip("#")
df
```

```python
df["v_conj"] = df.v_conj.str.strip("#")
df
```

```python
df['v_conj'] = df.v_conj.apply(lambda x: '$' + x + '#')
```
